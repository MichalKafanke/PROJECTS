```python
def min_max_function(x):
    min_number = min(x)
    max_number = max(x)
    print(f'Min: {min_number} \nMax: {max_number}')
    
x = [1, 4, -4, 7]
min_max_function(x)
```

# ZADANIE 3

```python
def counter_function(S):
    print(f'Amount of words: {len(S.split())}')
    S = S.lower()
    S = S.replace(' ', '')
    print(f'Amount of letters: {len(S)}')
    frequency_of_letters = {}
    for char in S:
        if char not in frequency_of_letters:
            frequency_of_letters[char] = 1
        else:
            frequency_of_letters[char] += 1
    for key, value in frequency_of_letters.items():
        print(f'Amount of letter: {key} - {value}')
    
S = 'Ala ma kota'
counter_function(S)
```

```python
def even_or_odd(x):
    if x % 2 == 0:
        print(f'Number {x} is even')
    else:
        print(f'Number {x} is odd')

even_or_odd(7)
even_or_odd(8)
```

# ZADANIE 5

```python
import string
import random

def generate_password():
    password = ''
    chars = string.ascii_letters + string.digits + string.punctuation
    for _ in range(17):
        password += password.join(random.choice(chars))
    return password

password = generate_password()
password
```

# ZADANIE 6

```python
def converter(number):
    if number == 1:
        km = float(input('Insert value in KM: '))
        print(f'{km} KM = {km * 0.62} MILES')
    elif number == 2:
        celsius = float(input('Insert value in CELSIUS: '))
        print(f'{celsius} CELSIUS = {celsius * 2 + 30} FAHRENHEIT')
    else:
        print('Incorrect value')

number = int(input('What do you want to convert (Choose number 1 or 2)?\n1. KM -> MILES\n2. CELSIUS -> FAHRENHEIT\n'))
converter(number)
```

# ZADANIE 7

```python
def bubble_sort(x):
    for _ in range(len(x)):
        for i in range(len(x) - 1):
            if x[i] > x[i + 1]:
                tmp = x[i]
                x[i] = x[i + 1]
                x[i + 1] = tmp
    print(f'Sorted list: {x}')

x = [23, 2, 7, 4, 5, 19, 33, 3, 2, 1, 17, 14, 27]
bubble_sort(x)
```
